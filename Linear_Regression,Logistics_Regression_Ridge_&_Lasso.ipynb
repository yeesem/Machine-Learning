```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df = pd.read_csv('https://raw.githubusercontent.com/yeesem/Machine-Learning-Datasets/main/Boston%20House%20Pricing.csv', header=None, delimiter=r"\s+", names=column_names)
```

```python
df.head()
```

```python
df.rename(columns = {'MEDV':'Price'},inplace = True)
```

```python
df.head()
```

```python
#Dividing the dataset into independent and dependent features
X = df.iloc[:,:-1]
y = df['Price'].iloc[:]
```

```python
X.head()
```

```python
y.shape
```

```python
#Test set = 33%,Train set = 77%
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size = 0.33,random_state = 42
)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
mse = cross_val_score(lin_reg,X_train,y_train,scoring = "neg_mean_squared_error",cv = 5)
print(mse)
mean_mse = np.mean(mse)
print(mean_mse)
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()

params = {'alpha' : [1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,40,45,50,55,100,150,180,200]}

ridge_regressor = GridSearchCV(ridge,params,scoring = 'neg_mean_squared_error',cv = 10)
ridge_regressor.fit(X_train,y_train)
```

```python
#Improve from -37 to -29
#If thes best score does not smaller than prev
#Ridge Regressor does not help [Try lasso regressor]
#Ridge Regressor reduce overfitting
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()

params = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,40,45,50,55,100]}

lasso_regressor = GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X_train,y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
from sklearn.metrics import r2_score

y_pred = ridge_regressor.predict(X_test)
r2Score = r2_score(y_pred,y_test)
```

```python
print(r2Score)
```

```python
y_pred = lasso_regressor.predict(X_test)
r2Score = r2_score(y_pred,y_test)
```

```python
print(r2Score)
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
```

```python
df['feature_names']
```

```python
df['data']
```

```python
X = pd.DataFrame(df['data'],columns = df['feature_names'])
```

```python
###Independent features
X.head()
```

```python
###Dependent feature
y = pd.DataFrame(df['target'],columns = ['Target'])
y.head()
```

```python
###Balance dataset
y['Target'].value_counts()
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split

XTrain,XTest,yTrain,yTest = train_test_split(
    X,y,test_size = 0.33,random_state = 42
)
```

```python
params = [{'C':[1,5,10]},{'max_iter' : [100,150]}]
```

```python
model1 = LogisticRegression(C = 100,max_iter = 100)
```

```python
model = GridSearchCV(model1,param_grid=params,scoring='f1',cv = 5)
```

```python
model.fit(XTrain,yTrain)
```

```python
print(model.best_params_)
print(model.best_score_)
```

```python
y_pred = model.predict(XTest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_pred,yTest)
```

```python
accuracy_score(yTest,y_pred)
```

```python
print(classification_report(y_pred,yTest))
```
